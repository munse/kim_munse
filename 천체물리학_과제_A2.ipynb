```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open('C:\\Users\\user\\Desktop\\새 폴더\\exoplanet.eu_catalog.csv','r', encoding = 'utf-8')
```

```python
df = pd.read_csv(f)
print(df.columns)
```

```python
name = df['# name']
semi_major_axis = df['semi_major_axis']
orbital_period = df['orbital_period']
detection = df['detection_type']
```

```python
st_RV = []
st_Im = []
st_PT = []
st_Ti = []
st_As = []
st_De = []
st_TTV = []
st_Mi = []

RV = []
Im = []
PT = []
Ti = []
As = []
De = []
TTV = []
Mi = []
```

```python
for i in range(0,4378):
    if detection[i] == 'Radial Velocity':
        RV.append(semi_major_axis[i])
        st_RV.append(orbital_period[i])
    if detection[i] == 'Imaging':
        Im.append(semi_major_axis[i])
        st_Im.append(orbital_period[i])
    if detection[i] == 'Primary Transit':
        PT.append(semi_major_axis[i])
        st_PT.append(orbital_period[i])
    if detection[i] == 'Timing':
        Ti.append(semi_major_axis[i])
        st_Ti.append(orbital_period[i])
    if detection[i] == 'Astrometry':
        As.append(semi_major_axis[i])
        st_As.append(orbital_period[i])
    if detection[i] == 'Default':
        De.append(semi_major_axis[i])
        st_De.append(orbital_period[i])
    if detection[i] == 'TTV':
        TTV.append(semi_major_axis[i])
        st_TTV.append(orbital_period[i])
    if detection[i] == 'Microlensing':
        Mi.append(semi_major_axis[i])
        st_Mi.append(orbital_period[i])
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.plot(st_RV,RV,'.',label = 'Radial Velocity')
plt.plot(st_Im,Im,'.',label = 'Imaging')
plt.plot(st_PT,PT,'.',label = 'Primary Transit')
plt.plot(st_Ti,Ti,'.',label = 'Timing')
plt.plot(st_As,As,'.',label = 'Astrometry')
plt.plot(st_De,De,'.',label = 'Default')
plt.plot(st_TTV,TTV,'.',label = 'TTV')
plt.plot(st_Mi,Mi,'.',label = 'Microlensing')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('orbital_period')
plt.ylabel('semi_major_axis')
plt.title('semi_major_axis vs orbital_period')
```
