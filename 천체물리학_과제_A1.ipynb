```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open('C:\\Users\\user\\Desktop\\새 폴더\\exoplanet.eu_catalog.csv','r', encoding = 'utf-8')
```

```python
df = pd.read_csv(f)
```

```python
name = df['# name']
mass = df['mass']
star_mass = df['star_mass']
detection = df['detection_type']
```

```python
st_RV = []
st_Im = []
st_PT = []
st_Ti = []
st_As = []
st_De = []
st_TTV = []
st_Mi = []

RV = []
Im = []
PT = []
Ti = []
As = []
De = []
TTV = []
Mi = []
```

```python
for i in range(0,4378):
    if detection[i] == 'Radial Velocity':
        RV.append(mass[i])
        st_RV.append(star_mass[i])
    if detection[i] == 'Imaging':
        Im.append(mass[i])
        st_Im.append(star_mass[i])
    if detection[i] == 'Primary Transit':
        PT.append(mass[i])
        st_PT.append(star_mass[i])
    if detection[i] == 'Timing':
        Ti.append(mass[i])
        st_Ti.append(star_mass[i])
    if detection[i] == 'Astrometry':
        As.append(mass[i])
        st_As.append(star_mass[i])
    if detection[i] == 'Default':
        De.append(mass[i])
        st_De.append(star_mass[i])
    if detection[i] == 'TTV':
        TTV.append(mass[i])
        st_TTV.append(star_mass[i])
    if detection[i] == 'Microlensing':
        Mi.append(mass[i])
        st_Mi.append(star_mass[i])
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.plot(st_RV,RV,'.',label = 'Radial Velocity')
plt.plot(st_Im,Im,'.',label = 'Imaging')
plt.plot(st_PT,PT,'.',label = 'Primary Transit')
plt.plot(st_Ti,Ti,'.',label = 'Timing')
plt.plot(st_As,As,'.',label = 'Astrometry')
plt.plot(st_De,De,'.',label = 'Default')
plt.plot(st_TTV,TTV,'.',label = 'TTV')
plt.plot(st_Mi,Mi,'.',label = 'Microlensing')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('star_mass')
plt.ylabel('mass')
plt.title('mass vs star mass')
```
